## Simple Linear Regression

Import the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Import the dataset

```python
dataset=pd.read_csv("Salary_Data.csv")
dataset.head()
```

```python
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

Splitting into train and test datas

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=0)
```

Training linear regression model

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

Predicting Test set

```python
y_pred=regressor.predict(x_test)
```

```python
plt.scatter(x_train,y_train,color="green")
plt.plot(x_train,regressor.predict(x_train),color="red")
plt.title("Salary vs Exp Training set")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

Visualising test set results

```python
plt.scatter(x_test,y_test,color="green")
plt.plot(x_test,regressor.predict(x_test),color="red")
plt.title("Salary vs Exp Test set")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

Evaluating the model for the dataset

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
m=dataset["Salary"].mean()
```

```python
m1= mean_absolute_error(y_test,y_pred)#checks general error
```

```python
m2= np.sqrt(mean_squared_error(y_test,y_pred))#check for outliers
```

```python
Accuracy=(m-m1)/m
```
